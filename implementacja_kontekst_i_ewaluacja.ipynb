```python
import pandas as pd
import requests
import requests
import Levenshtein
from transformers import pipeline
```

```python
answers = requests.get('https://raw.githubusercontent.com/pochu1337/qa/main/Odpowiedzi')
context = requests.get('https://raw.githubusercontent.com/pochu1337/qa/main/context')
questions = requests.get('https://raw.githubusercontent.com/pochu1337/qa/main/Pytania')
questions = questions.text.split('\n')[:200]
context = context.text.split('\n')[:200]
answers = answers.text.split('\n')[:200]
```

```python
qa_pipeline = pipeline(
    "question-answering",
    model="henryk/bert-base-multilingual-cased-finetuned-polish-squad2",
    tokenizer="henryk/bert-base-multilingual-cased-finetuned-polish-squad2"
)

dane = qa_pipeline({
    'context': context,
    'question': questions})
```

```python
klucz = requests.get('https://raw.githubusercontent.com/pochu1337/qa/main/Odpowiedzi')
klucz = klucz.text.lower().split('\n')
```

```python
dane = [dana['answer'].lower() for dana in dane]
klucz = klucz[:200]
```

```python

def lev(dane):
  ilegit = []
  ok = []
  distances = []
  for expect,ours in zip(klucz,dane):
    distances.append((Levenshtein.distance(ours,expect))/(len(expect)/2))
    if (Levenshtein.distance(ours,expect))/(len(expect)/2) <= 0.6:
      ilegit.append(1)
      ok.append(1)
    else:
      ok.append(0)
  return ilegit, ok, distances
```
